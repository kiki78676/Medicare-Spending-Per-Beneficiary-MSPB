```python
import urllib
from sqlalchemy import create_engine

odbc_str = (
    "Driver={ODBC Driver 18 for SQL Server};"
    "Server=DRGAMER123\\SQLEXPRESS;"
    "Database=Medicare_Hospital_db;"
    "Trusted_Connection=yes;"
    "Encrypt=no;"
)

params = urllib.parse.quote_plus(odbc_str)

engine = create_engine(
    f"mssql+pyodbc:///?odbc_connect={params}",
    fast_executemany=True 
)

print("Engine ready!")
```

```python
import pandas as pd
csv_path = r"C:\Users\kisho\Downloads\Big projects\cool\Hospital\CSV Files\mspb_clean.csv"
df = pd.read_csv(csv_path, low_memory=False)

print(df.shape)    
print(df.columns[:10]) 

# 2. Push into SQL Server
df.to_sql(
    "mspb_clean", 
    con=engine,
    if_exists="append",    
    index=False
)

print("Loaded CSV into SQL Server!")
```
